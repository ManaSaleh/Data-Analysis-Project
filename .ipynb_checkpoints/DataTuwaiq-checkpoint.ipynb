**Data Standardization**

```python
def convert_score(score, score_system):
    if score_system == 4:
        return score * (5 / 4)
    elif score_system == 100:
        return score / 20
    else:
        return score
df_train['GPA'] = df_train.apply(lambda row: convert_score(row['University Degree Score'], row['University Degree Score System']), axis=1)
df_train.sample(20)
```

```python
df_train_new = pd.read_csv('new-dataset-train.csv')
df_test_new = pd.read_csv('new-dataset-test.csv')
```

```python
# Label Encoding for Gender column
label_encoder = LabelEncoder()
df_train_new['Gender'] = label_encoder.fit_transform(df_train_new['Gender'])
df_train_new.head()
```

```python
# Label Encoding for Gender column
label_encoder = LabelEncoder()
df_test_new['Gender'] = label_encoder.fit_transform(df_test_new['Gender'])
df_test_new.head() 
```

```python
# Get unique values in the "Program Skill Level" column
unique_values = df_train_new['Program Skill Level'].unique()
# Define the categories for ordinal encoding
ordinal_categories = [unique_values]
# Now, apply ordinal encoding for Program Skill Level
ordinal_encoder = OrdinalEncoder(categories=ordinal_categories)
df_train_new['Program Skill Level'] = ordinal_encoder.fit_transform(df_train_new[['Program Skill Level']])
df_train_new[['Program Skill Level']].sample(10)
```

```python
# Get unique values in the "Program Skill Level" column
unique_values = df_test_new['Program Skill Level'].unique()
# Define the categories for ordinal encoding
ordinal_categories = [unique_values]
# Now, apply ordinal encoding for Program Skill Level
ordinal_encoder = OrdinalEncoder(categories=ordinal_categories)
df_test_new['Program Skill Level'] = ordinal_encoder.fit_transform(df_test_new[['Program Skill Level']])
df_test_new[['Program Skill Level']].sample(10)
```

```python
columns_to_encode = ['Home Region', 'Program Main Category Code', 'Program Sub Category Code',
                     'Technology Type', 'Program Presentation Method', 'Completed Degree', 'Level of Education',
                     'Education Speaciality', 'Employment Status', 'Job Type', 'Still Working']
for col in columns_to_encode:
    label_encoder = LabelEncoder()
    df_train_new[col] = label_encoder.fit_transform(df_train_new[col])
```

```python
columns_to_encode = ['Home Region', 'Program Main Category Code', 'Program Sub Category Code',
                     'Technology Type', 'Program Presentation Method', 'Completed Degree', 'Level of Education',
                     'Education Speaciality', 'Employment Status', 'Job Type', 'Still Working']
for col in columns_to_encode:
    label_encoder = LabelEncoder()
    df_test_new[col] = label_encoder.fit_transform(df_test_new[col])
```

```python
date_columns = ['Program Start Date', 'Program End Date']
df_train_new[date_columns] = df_train_new[date_columns].apply(pd.to_datetime)
# Extract relevant features
df_train_new['Program Start Year'] = df_train_new['Program Start Date'].dt.year
df_train_new['Program Start Month'] = df_train_new['Program Start Date'].dt.month
df_train_new['Program Start Day'] = df_train_new['Program Start Date'].dt.day
df_train_new['Program End Year'] = df_train_new['Program End Date'].dt.year
df_train_new['Program End Month'] = df_train_new['Program End Date'].dt.month
df_train_new['Program End Day'] = df_train_new['Program End Date'].dt.day
# Drop the original date columns if no longer needed
df_train_new.drop(columns=date_columns, inplace=True)
df_train_new.head()
```

```python
date_columns = ['Program Start Date', 'Program End Date']
df_test_new[date_columns] = df_test_new[date_columns].apply(pd.to_datetime)
# Extract relevant features
df_test_new['Program Start Year'] = df_test_new['Program Start Date'].dt.year
df_test_new['Program Start Month'] = df_test_new['Program Start Date'].dt.month
df_test_new['Program Start Day'] = df_test_new['Program Start Date'].dt.day
df_test_new['Program End Year'] = df_test_new['Program End Date'].dt.year
df_test_new['Program End Month'] = df_test_new['Program End Date'].dt.month
df_test_new['Program End Day'] = df_test_new['Program End Date'].dt.day
# Drop the original date columns if no longer needed
df_test_new.drop(columns=date_columns, inplace=True)
df_test_new.head()
```

```python
# Select numerical columns to scale
numerical_columns = ['Age', 'Program Days', 'GPA']
# Initialize MinMaxScaler
scaler = MinMaxScaler()
# Fit scaler on the numerical data and transform it
df_train_new[numerical_columns] = scaler.fit_transform(df_train_new[numerical_columns])
df_train_new.head()
```

```python
# Select numerical columns to scale
numerical_columns = ['Age', 'Program Days', 'GPA']
# Initialize MinMaxScaler
scaler = MinMaxScaler()
# Fit scaler on the numerical data and transform it
df_test_new[numerical_columns] = scaler.fit_transform(df_test_new[numerical_columns])
df_test_new.head()
```

```python
data = merged_data
# Replace missing values in 'Y' with a placeholder value (e.g., -1)
data['Y'].fillna(-1, inplace=True)
# Splitting features (X) and target variable (y)
X = data.drop(columns=['Y'])  # Assuming 'Y' is the target variable
y = data['Y']
# Splitting the dataset into training and testing sets, stratifying by y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
train_y_counts = y_train.value_counts(normalize=True)
test_y_counts = y_test.value_counts(normalize=True)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.pie(train_y_counts, labels=train_y_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Target Variable (Y) in Training Set')
plt.subplot(1, 2, 2)
plt.pie(test_y_counts, labels=test_y_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Target Variable (Y) in Testing Set')
plt.tight_layout()
```

```python
# Separate instances with non-missing 'Y' values for training
train_data = data[data['Y'] != -1]
X_train = train_data.drop(columns=['Y'])
y_train = train_data['Y']
```

```python
# Separate instances with missing 'Y' values for testing
test_data = data[data['Y'] == -1]
X_test = test_data.drop(columns=['Y'])
y_test = test_data['Y']
```

```python
# Drop 'Student ID' and 'Program ID' columns from X_train and X_test
X_train.drop(columns=['Student ID', 'Program ID'], inplace=True)
X_test.drop(columns=['Student ID', 'Program ID'], inplace=True)
```

```python
# Instantiate classifiers
catboost_clf = CatBoostClassifier()    # CatBoostClassifier instance

# Train CatBoostClassifier
catboost_clf.fit(X_train, y_train)

# Make predictions
catboost_preds = catboost_clf.predict(X_test)  # Predictions using CatBoostClassifier
```

```python
print("Length of X_test:", len(X_test))
print("Length of catboost_preds:", len(catboost_preds))
```
